```python
import os
# Assuming you have a .env file and use python-dotenv
# from dotenv import load_dotenv
# load_dotenv()

from langchain_google_genai import ChatGoogleGenerativeAI

# Define configuration variables clearly
# LangChain will automatically look for the GOOGLE_API_KEY environment variable.
my_model_name = "gemini-2.0-flash"
my_temperature = 0.7 
my_max_tokens = 2048 

# Initialize the model
# No explicit API key needed if the environment variable is set
llm = ChatGoogleGenerativeAI(
    model=my_model_name,google_api_key = ""
    temperature=my_temperature,
    max_output_tokens=my_max_tokens
)

# You can now use the 'llm' object to invoke the model
response = llm.invoke("Your prompt here")
```

```python
from langchain_groq import ChatGroq

# The ChatGroq class automatically finds the GROQ_API_KEY environment variable
my_api_key = ""

llm = ChatGroq(
    model="openai/gpt-oss-20b",
    temperature=0.7,
    groq_api_key=my_api_key  # Pass the key explicitly
)

response = llm.invoke("are you free")
print(response.content)
```
